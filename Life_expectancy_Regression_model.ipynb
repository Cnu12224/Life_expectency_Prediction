```python
import pandas as pd
df = pd.read_csv('Life Expectancy Data.csv')
```

```python
df.info()
```

```python
df_encoded = pd.get_dummies(df, columns=['Country', 'Status'])
df_encoded = df_encoded.fillna(0).astype(int)
display(df_encoded.head())
```

```python
corr = df_encoded.corr()
display(corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(corr)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Calculate the absolute correlation of all columns with 'Life expectancy'
abs_corr_with_life_expectancy = corr['Life expectancy '].abs()

# Identify columns to keep (absolute correlation >= 0.4) and exclude 'Life expectancy' itself
# Also exclude all 'Country_' columns
columns_to_keep = abs_corr_with_life_expectancy[abs_corr_with_life_expectancy >= 0.4].index.tolist()
columns_to_keep = [col for col in columns_to_keep if col != 'Life expectancy ' and not col.startswith('Country_')]

# Add 'Life expectancy' back to the list of columns to keep, as it's the target variable
columns_to_keep.append('Life expectancy ')

# Filter the DataFrame to keep only the selected columns
df_filtered = df_encoded[columns_to_keep]

display(df_filtered.head())
```

```python
# Identify all 'Country_' columns in the original encoded DataFrame
country_columns = [col for col in df_encoded.columns if col.startswith('Country_')]

# Select these columns from the original encoded DataFrame
df_country_columns = df_encoded[country_columns]

# Concatenate the filtered DataFrame and the country columns DataFrame
df_filtered_with_country = pd.concat([df_filtered, df_country_columns], axis=1)

display(df_filtered_with_country.head())
```

```python
df.info()
```

```python
# Fill null values with the median of each column
df_filled_median = df_filtered_with_country.fillna(df_filtered_with_country.median())

# Display the info to check for null values
display(df_filled_median.info())
```

## Prepare data for modeling

### Subtask:
Separate the features (X) and the target variable (y) from the `df_filled_median` DataFrame.


**Reasoning**:
Separate the features (X) and the target variable (y) from the `df_filled_median` DataFrame as instructed.



```python
X = df_filled_median.drop('Life expectancy ', axis=1)
y = df_filled_median['Life expectancy ']

display(X.head())
display(y.head())
```

## Define the model

### Subtask:
Initialize a linear regression model.


**Reasoning**:
Initialize a linear regression model.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

## Perform cross-validation

### Subtask:
Use K-Fold cross-validation (with 5 splits) on the data.


**Reasoning**:
Perform K-Fold cross-validation on the data using the initialized linear regression model.



```python
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X, y, cv=kf)

print("Cross-validation scores:", cv_scores)
```

## Evaluate the model

### Subtask:
Calculate and print appropriate regression metrics (e.g., Mean Squared Error, R-squared) from the cross-validation results.


**Reasoning**:
Calculate and print the mean and standard deviation of the cross-validation R-squared scores and the mean of the cross-validation Mean Squared Errors.



```python
print(f"Mean R-squared: {cv_scores.mean():.4f}")
print(f"Standard Deviation of R-squared: {cv_scores.std():.4f}")

# Calculate MSE using cross_val_score with scoring='neg_mean_squared_error'
mse_scores = cross_val_score(model, X, y, cv=kf, scoring='neg_mean_squared_error')
mean_mse = -mse_scores.mean()
print(f"Mean Squared Error: {mean_mse:.4f}")
```

## Summary:

### Data Analysis Key Findings

*   A Linear Regression model was initialized.
*   5-fold cross-validation was performed on the data using the Linear Regression model.
*   The mean R-squared from the cross-validation was 0.8259.
*   The standard deviation of the R-squared scores was 0.0380.
*   The mean Mean Squared Error from the cross-validation was 18.7649.

### Insights or Next Steps

```python
import numpy as np

# Identify columns to transform (all columns except country and status columns)
columns_to_transform = [col for col in df_filled_median.columns if not (col.startswith('Country_') or col.startswith('Status_'))]

# Apply log transformation (adding a small constant to handle zeros)
for col in columns_to_transform:
    # Add a small constant to handle zero values before log transformation
    df_filled_median[col] = np.log1p(df_filled_median[col])

# Display histograms for the transformed columns to check the effect of transformation
for col in columns_to_transform:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df_filled_median, x=col, kde=True)
    plt.title(f'Histogram of {col} (Transformed)')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Define a function to perform IQR treatment and replace outliers with median
def iqr_outlier_treatment(df, columns, threshold=1.5):
    df_treated = df.copy()
    for col in columns:
        Q1 = df_treated[col].quantile(0.25)
        Q3 = df_treated[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        # Replace outliers with the median
        median_val = df_treated[col].median()
        df_treated[col] = np.where(
            (df_treated[col] < lower_bound) | (df_treated[col] > upper_bound),
            median_val,
            df_treated[col]
        )
    return df_treated

# Identify columns to apply IQR treatment (all columns except country and status columns)
columns_to_treat = [col for col in df_filled_median.columns if not (col.startswith('Country_') or col.startswith('Status_'))]

# Apply IQR treatment to the specified columns
df_treated_iqr = iqr_outlier_treatment(df_filled_median.copy(), columns_to_treat)

# Display info and head of the treated DataFrame
display(df_treated_iqr.info())
display(df_treated_iqr.head())
```

```python
# Identify columns to plot (all columns except country and status columns)
columns_to_plot = [col for col in df_treated_iqr.columns if not (col.startswith('Country_') or col.startswith('Status_'))]

# Plot histograms for the treated columns
for col in columns_to_plot:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df_treated_iqr, x=col, kde=True)
    plt.title(f'Histogram of {col} (IQR Treated)')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
